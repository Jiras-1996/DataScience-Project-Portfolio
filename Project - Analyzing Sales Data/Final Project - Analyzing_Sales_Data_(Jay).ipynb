### Q4.Your friend ask for all order data in `California` and `Texas` in `2017` (look at Order Date), send him csv file

### Q5.How much `total sales, average sales, and standard deviation` of sales your company make in `2017`?

```python
# filter sale data in 2017
df_2017 = df[ df['Order Date'].dt.year == 2017 ]

df_2017.head()
```

```python
# Calculate the Statistical Values of Sales
df_agg_2017 = df_2017['Sales'].agg(['sum', 'mean', 'std']).reset_index()

# Assign column name header
df_agg_2017.columns = ['Statistical Values', 'Sales']

df_agg_2017
```

### Q7.Which `top 5 States` have the least total sales between `15 April 2019 - 31 December 2019`?

```python
# filter data with "Order Date" in range 15 April 2019 - 31 December 2019
df_date = df[ (df['Order Date'] > '04-15-2019') & (df['Order Date'] < '12-31-2019') ]

df_date.head()
```

```python
# Calculate total sales in all state
df_total_sales = df_date.groupby('State')['Sales'].sum().reset_index()

df_total_sales
```

```python
# sorting data to find Top 5 states with "Sales"
top5_states = df_total_sales.sort_values('Sales').head(5)

# assign column name header
top5_states.columns = ['State', 'Total Sales']

top5_states
```

### Q8.What is the `proportion of total sales (%)` in West + Central in `2019` e.g. 25%?
- Filter the specific Order Date and Regions
- Calculate the proportion
- Display the result to %

```python
# filter data with "Order Date" = 2019 and "Region" = West + Central
df_2019_WC = df[ df['Order Date'].dt.year == 2019 ].query(" Region == 'West' | Region == 'Central' ")

df_2019_WC.head()
```

```python
# calculate proportion of total sales
df_result = ( df_2019_WC['Sales'].sum() / df['Sales'].sum() )

# Display the result to %
result = f"The proportion of total sales in West + Centra in 2019 is {df_result:.0%}" #.0% used convert to percent

print(result)
```

### Q9. Find top `10 popular product`s in terms of number of orders vs. total sales during `[2019-2020]`
- Filter the specific interval
- Calculate the sum of Quantity and Sales grouped by Product Name (separate to 2 objects)
- Compare 2 objects

```python
# filter data in range years 2019-2020 with "Order Date"
df_years = df[ (df['Order Date'] >= '01-01-2019') & (df['Order Date'] <= '31-12-2020') ].sort_values('Order Date')

df_years.head()
```

```python
# calculate sum of Quantitiy
top10_orders = df_years.groupby('Product Name')['Quantity'].agg( ['sum'] ).sort_values('sum', ascending=False).head(10).reset_index()

# assign column names
top10_orders.columns = ['Product Name', 'Number of orders']
```

```python
# calculate sum of Sales
top10_sales = df_years.groupby('Product Name')['Sales'].agg( ['sum'] ).sort_values('sum', ascending=False).head(10).reset_index()

# assign column names
top10_sales.columns = ['Product Name', 'Total sales']
```

```python
# compare 2 objects
print(top10_orders)
print("\n")
print(top10_sales)
```
